## PROBLEM 1
### Find the total number of crimes recorded in the CRIME table.


```python
%sql SELECT COUNT(*) AS TOTAL_CRIMES FROM CHICAGO_CRIME_DATA;
```

## PROBLEM 2
### List community areas with per capita income less than 11000.

```python
%sql SELECT COMMUNITY_AREA_NAME, PER_CAPITA_INCOME FROM CHICAGO_CENSUS_DATA WHERE PER_CAPITA_INCOME < 11000;
```

## PROBLEM 3
### List all case numbers for crimes involving minors?

```python
%sql SELECT CASE_NUMBER FROM CHICAGO_CRIME_DATA WHERE PRIMARY_TYPE = 'OFFENSE INVOLVING CHILDREN';
```

## PROBLEM 4
### List all kidnapping crimes involving a child?

```python
%sql SELECT PRIMARY_TYPE,CASE_NUMBER FROM CHICAGO_CRIME_DATA WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%';
```

## PROBLEM 5
### What kind of crimes were recorded at schools?

```python
%sql SELECT NAME_OF_SCHOOL, PRIMARY_TYPE \
FROM CHICAGO_CRIME_DATA CD, CHICAGO_PUBLIC_SCHOOLS PB \
WHERE CD.COMMUNITY_AREA_NUMBER = PB.COMMUNITY_AREA_NUMBER LIMIT 5;
```

## PROBLEM 6
### List the average safety score for all types of schools.

```python
%sql SELECT AVG(SAFETY_SCORE) AS AVE_SAFETY_SCORE, \"ELEMENTARY, MIDDLE or HIGHSCHOOL\" \
FROM CHICAGO_PUBLIC_SCHOOLS \
GROUP BY \'ELEMENTARY, MIDDLE or HIGHSCHOOL\';
```

## PROBLEM 7
### List 5 community areas with highest % of households below poverty line.

```python
%sql SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY \
FROM CHICAGO_CENSUS_DATA \
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC NULLS LAST LIMIT 5;
```

## PROBLEM 8
### Which community area(number) is most crime prone?

```python
%sql SELECT COMMUNITY_AREA_NUMBER, COUNT(*) AS NUM_CRIME \
FROM CHICAGO_CRIME_DATA \
GROUP BY COMMUNITY_AREA_NUMBER ORDER BY NUM_CRIME DESC LIMIT 1;
```

## PROBLEM 9
### Use a sub-query to find the name of the community area with highest hardship index.

```python
%sql SELECT COMMUNITY_AREA_NAME \
FROM CHICAGO_CENSUS_DATA \
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CHICAGO_CENSUS_DATA);
```

## PROBLEM 10
### Use a sub-query to determine the Community Area Name with most number of crimes?

```python
%sql CREATE VIEW total_crime AS SELECT COMMUNITY_AREA_NUMBER, COUNT(*) AS number_of_crime FROM CHICAGO_CENSUS_DATA GROUP BY COMMUNITY_AREA_NUMBER; 
```
